```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.ticker as ticker
import numpy as np
%matplotlib inline
```

```python
models = ['conv5_block3', 'conv5_block2', 'conv5_block1', 'conv4_block6', 'conv4_block5', 'conv4_block4', 'conv4_block3', 'conv4_block2', 'conv4_block1', 
          'conv3_block4', 'conv3_block3', 'conv3_block2', 'conv3_block1', 'conv2_block2', 'conv2_block1', 'CNN-6']
# models = ['c5b3 (resnet50)', 'conv5_block2', 'c5b1', 'c4b6', 'c4b5', 'c4b4', 'c4b3', 'c4b2', 'c4b1', 
#           'c3b4', 'c3b3', 'c3b2', 'c3b1', 'c2b2', 'c2b1']
params = [23542788, 19077124, 14652292, 8593284, 7471492, 6325252, 5206532, 4087812, 2969092, 
         1452036, 1171204, 890372, 609540, 157060, 86276, 102828580]
```

```python
fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(1, 1, 1)

ax.set_ylim((0, 105000000))
ax.set_xlabel('Resnet50 variations')
ax.set_ylabel('Model Parameters')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.plot(models, params, marker='o', markevery=[0, 1, 2, 3, 5, 7, 8, 9, 11, 12, 13, 14])

ylabels = ['{:,.0f}'.format(x) + 'M' for x in ax.get_yticks()/1000000]
ax.set_yticklabels(ylabels)

for tick in ax.get_xticklabels():
    tick.set_rotation(90)
    
# ax.text(models[1], params[1], params[1], horizontalalignment='left')
ann_params = ['{:,.2f}'.format(x/1000000) + 'M' for x in params]

ann0 = '%s (Resnet50)' % ann_params[0]
ax.annotate(ann0, (models[0], params[0]), textcoords='offset points', xytext=(10,-3), ha='left')
ax.annotate(ann_params[1], (models[1], params[1]), textcoords='offset points', xytext=(10,-3), ha='left')
ax.annotate(ann_params[2], (models[2], params[2]), textcoords='offset points', xytext=(10,-3), ha='left')
ax.annotate(ann_params[3], (models[3], params[3]), textcoords='offset points', xytext=(-10,-3), ha='right')
ax.annotate(ann_params[4], (models[4], params[4]), textcoords='offset points', xytext=(-4,7), ha='left')
ax.annotate(ann_params[6], (models[6], params[6]), textcoords='offset points', xytext=(-4,7), ha='left')

ax.annotate(ann_params[10], (models[10], params[10]), textcoords='offset points', xytext=(0,7), ha='center')

ax.scatter([models[4], models[6], models[10]], [params[4], params[6], params[10]], marker='o', color='r')
```

```python
import math

fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(1, 1, 1)

ax.set_ylim((10, 20))
ax.set_xlabel('Model Architectures')
ax.set_ylabel('Number of Model Parameters (log scale)')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

p = [math.log(x) for x in params]

ax.plot(models[:-1], p[:-1], marker='o', markevery=[0, 1, 2, 3, 5, 7, 8, 9, 11, 12, 13, 14], label='ResNet-50 variations')
ax.plot(models[-1:], p[-1:], marker='X', label='Yu et al CNN-6')

for tick in ax.get_xticklabels():
    tick.set_rotation(90)
    
ann_params = ['{:,.2f}'.format(x/1000000) + 'M' for x in params]
ann_params_k = ['{:,.2f}'.format(x/1000) + 'K' for x in params]

ann0 = 'ResNet-50\n%s' % ann_params[0]
ax.annotate(ann0, (models[0], p[0]), textcoords='offset points', xytext=(-10,6), ha='left')
# ax.annotate(ann_params[1], (models[1], p[1]), textcoords='offset points', xytext=(10,-3), ha='left')
# ax.annotate(ann_params[2], (models[2], p[2]), textcoords='offset points', xytext=(10,-3), ha='left')
# ax.annotate(ann_params[3], (models[3], p[3]), textcoords='offset points', xytext=(-10,-3), ha='right')

ax.annotate(ann_params[4], (models[4], p[4]), textcoords='offset points', xytext=(-4,7), ha='left')
ax.annotate(ann_params[6], (models[6], p[6]), textcoords='offset points', xytext=(-4,7), ha='left')
ax.annotate(ann_params[10], (models[10], p[10]), textcoords='offset points', xytext=(0,7), ha='center')

ax.annotate(ann_params_k[-2], (models[-2], p[-2]), textcoords='offset points', xytext=(0,-20), ha='center')

ax.annotate(ann_params[-1], (models[-1], p[-1]), textcoords='offset points', xytext=(0,7), ha='center')

ax.scatter([models[4], models[6], models[10]], [p[4], p[6], p[10]], marker='s')

ax.annotate('Top Performers', (models[4], p[4]), textcoords='offset points', xytext=(20,-80), ha='center')
ax.legend(loc=3)
```
